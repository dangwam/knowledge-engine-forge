```python
import pandas as pd
import os
import matplotlib.pyplot as plt

base_dir = r'C:\Users\pythonProject\test'
file_names = ['O.csv','SPY.csv','RIVN.csv','TSLA.csv','FSR.csv']
symbols = ['SPY','RIVN','TSLA','FSR']
dates = pd.date_range('2023-01-01','2023-08-30')
df1 = pd.DataFrame(index=dates)
sd = '03-01-2023'
ed = '03-15-2023'
title = 'Stock Prices'
# function to return complete file path to read_csv

def fpath(symbol,base_dir):
  return os.path.join(base_dir,"{}.csv".format(str(symbol)))

for symbol in symbols:
  df_temp = pd.read_csv(fpath(symbol,base_dir),index_col='Date',parse_dates=True,usecols = ['Date','Adj Close'],na_values='nan')
  df_temp = df_temp.rename(columns={'Adj Close':symbol}) 
  df1 = df1.join(df_temp)
  if symbol == 'SPY':
    df1 = df1.dropna(subset=["SPY"])
  df1 = df1.fillna(1)

''' Divide df1 by itself after the join() method is called
    Note that when you use a single bracket [] in a loc or iloc then the usage is loc[row:row,col] or iloc[row] and it returns you the cells . 
    But if you are using double brackets [[]] then the usage is [[columns]]
'''

if not df1.empty:
    df1 = df1 / df1.iloc[0]  # Use iloc to select the first row (index 0)

print(df1)
df1.to_csv(os.path.join(base_dir,"output.csv"))

#print (df1[['SPY','TSLA']])
#print(df1.loc['03-01-2023':'03-15-2023',['SPY','TSLA']])
#########################################################
#plots in pandas
############Normalizing the data dividing by low value

ax = df1.plot(title = title, fontsize=5)
ax.set_xlabel("Date")
ax.set_ylabel("Price")
plt.show()

'''
                        ROLLING STATISTICS - moving averages
'''
```

```python
'''
rolling mean using pandas
'''
import pandas as pd
import os
import matplotlib.pyplot as plt

base_dir = r'C:\Users\pythonProject\test'
file_names = ['SPY.csv']
symbol = 'SPY'
fpath=os.path.join(base_dir,"{}.csv".format(str(symbol)))
dates = pd.date_range('2023-01-01','2023-08-30')
df = pd.DataFrame(index=dates)

df_temp = pd.read_csv(fpath,index_col='Date',parse_dates=True,usecols = ['Date','Adj Close'],na_values='nan')
df_temp = df_temp.rename(columns={'Adj Close':symbol}) 
df = df.join(df_temp)
if symbol == 'SPY':
    df = df.dropna(subset=["SPY"])

ax = df['SPY'].plot(title='SPY Rolling mean',label = 'SPY')
rm = df.rolling(window=20,win_type='gaussian').mean(std=2)
rm.plot(title='Rolling mean',ax = ax)
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.legend(loc='upper left')
plt.show()

'''
calculating bollinger bands
1). calculate the rolling mean
2). calculate the standard deviation
3). calculate the upper and lower bounds
'''
```
